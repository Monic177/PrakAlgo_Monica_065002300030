```python
# Inisialisasi list kosong untuk menyimpan data mahasiswa
data_mahasiswa = []

# Fungsi untuk input data mahasiswa
def input_data():
    nama = input("Masukkan nama mahasiswa: ")
    nilai_praktikum = []
    for i in range(3):
        nilai = float(input(f"Masukkan nilai praktikum {i+1}: "))
        nilai_praktikum.append(nilai)
    data_mahasiswa.append({'Nama': nama, 'Praktikum': nilai_praktikum})

# Fungsi untuk view data dalam tabel
def view_data():
    print("Data Mahasiswa:")
    print("Nama\tPrak 1\tPrak 2\tPrak 3")
    for mahasiswa in data_mahasiswa:
        print(f"{mahasiswa['Nama']}\t{mahasiswa['Praktikum'][0]}\t{mahasiswa['Praktikum'][1]}\t{mahasiswa['Praktikum'][2]}")

# Fungsi untuk menghitung rata-rata praktikum tiap mahasiswa
def hitung_rata_mahasiswa():
    for mahasiswa in data_mahasiswa:
        rata_rata = sum(mahasiswa['Praktikum']) / len(mahasiswa['Praktikum'])
        print(f"Rata-rata praktikum {mahasiswa['Nama']}: {rata_rata:.2f}")

# Fungsi untuk menghitung rata-rata tiap praktikum
def hitung_rata_praktikum():
    total_praktikum = [0, 0, 0]
    for mahasiswa in data_mahasiswa:
        for i in range(3):
            total_praktikum[i] += mahasiswa['Praktikum'][i]
    rata_praktikum = [total / len(data_mahasiswa) for total in total_praktikum]
    print(f"Rata-rata praktikum:\nPrak 1: {rata_praktikum[0]:.2f}\nPrak 2: {rata_praktikum[1]:.2f}\nPrak 3: {rata_praktikum[2]:.2f}")

# Fungsi untuk mencari nilai rata-rata praktikum mahasiswa
def cari_rata_mahasiswa(nama_mahasiswa):
    for mahasiswa in data_mahasiswa:
        if mahasiswa['Nama'] == nama_mahasiswa:
            rata_rata = sum(mahasiswa['Praktikum']) / len(mahasiswa['Praktikum'])
            print(f"Rata-rata praktikum {nama_mahasiswa}: {rata_rata:.2f}")
            return
    print("Nama mahasiswa tidak ditemukan.")

# Fungsi untuk update nilai praktikum mahasiswa
def update_nilai():
    nama = input("Masukkan nama mahasiswa yang akan diupdate: ")
    for mahasiswa in data_mahasiswa:
        if mahasiswa['Nama'] == nama:
            nilai_praktikum = []
            for i in range(3):
                nilai = float(input(f"Masukkan nilai praktikum {i+1} yang baru: "))
                nilai_praktikum.append(nilai)
            mahasiswa['Praktikum'] = nilai_praktikum
            print("Nilai berhasil diupdate.")
            return
    print("Nama mahasiswa tidak ditemukan.")

# Loop program utama
while True:
    print("\nPilih operasi yang diinginkan:")
    print("1. Input data")
    print("2. View data")
    print("3. Hitung nilai rata-rata praktikum tiap mahasiswa")
    print("4. Hitung nilai rata-rata tiap praktikum")
    print("5. Mencari nilai rata-rata praktikum mahasiswa")
    print("6. Update nilai praktikum mahasiswa")
    print("7. Exit")

    pilihan = input("Masukkan pilihan (1-7): ")

    if pilihan == '1':
        input_data()
    elif pilihan == '2':
        view_data()
    elif pilihan == '3':
        hitung_rata_mahasiswa()
    elif pilihan == '4':
        hitung_rata_praktikum()
    elif pilihan == '5':
        nama = input("Masukkan nama mahasiswa: ")
        cari_rata_mahasiswa(nama)
    elif pilihan == '6':
        update_nilai()
    elif pilihan == '7':
        print("Terima kasih!")
        break
    else:
        print("Pilihan tidak valid. Silakan masukkan angka 1-7.")
```
